$ f(x) = cos(x)$

$[a, b] = [3.5, 4.5]$ 

**Точки интерполирования:** $\; \overline{x_1} = 3.45; \; \; \overline{x_3} = 3.95$                            

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
import pandas as pd
```

```python
def f(x):
    return np.cos(x)
```

```python
def args_values_arrays(xl, xr, h=0.1):
    x = xl
    values = []
    args = []
    while x <= xr:
        args.append(np.round(x, 1))
        values.append(np.round(f(x), 5))
        x += h
    return args, values
```

```python
args, values = args_values_arrays(3.5, 4.5)
```

```python
fig = plt.figure()
fig.set_size_inches(18.5, 10.5)

a = fig.add_subplot(1, 2, 1)
x = np.linspace(0, 10, 100)
y = [np.cos(i) for i in x]
a.plot(x, y, color='black')
a.hlines(0, -1, 11, color='black')
a.axvline(x=3.5, color='blue')
a.axvline(x=4.5, color='blue')
a.grid(True)

a = fig.add_subplot(1, 2, 2)
x = np.linspace(3.5, 4.5, 100)
y = [np.cos(i) for i in x]
a.plot(x, y, color='black')
a.axvline(x=3.5, color='blue')
a.axvline(x=4.5, color='blue')
a.grid(True)
```

## Таблица значений функции

```python
def fill_table(args, values, diffs=[], func='cos(x)'):
    a = {}
    a['x'] = args
    a[func] = values
    if len(diffs) > 0:
        i = 0
        for diff in diffs:
            key = 'delta' + str(i+1)
            a[key] = diffs[i]
            i += 1
    df = pd.DataFrame.from_dict(a, orient='index')
    df = df.transpose()
    return df.style.hide_index()
```

```python
fill_table(args, values)
```

## Конечные разницы

```python
def C(n, k):
    if 0 <= k <= n:
        nn = 1
        kk = 1
        for t in range(1, min(k, n - k) + 1):
            nn *= n
            kk *= t
            n -= 1
        return nn // kk
    else:
        return 0
```

```python
def finite_difference(values, rank=4, diffs=[]):
    i = 0
    diffs_for_one = []
    if len(diffs) < rank:
        for value in values:
            if (i < len(values) - 1):
                diffs_for_one.append(np.round(value - values[i+1], 5))
                i += 1
            else:
                diffs.append(diffs_for_one)
                finite_difference(diffs_for_one, rank, diffs)
    return diffs
```

```python
diffs  = finite_difference(values)
```

```python
fill_table(args, values, diffs)
```
